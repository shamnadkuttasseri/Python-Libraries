### basic plotting with pyplot


```python
#sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plotting a simple line graph
plt.plot(x,y)

#adding labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')

#display the plot 
plt.show()
```

#### 1.line plot

```python
#sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plotting a simple line graph
plt.plot(x,y, label='Line Plot')

#adding labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Graph')

#display the legend
plt.legend()

#display the plot 
plt.show()
```

#### 2.scatter plot

```python
#sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plotting a scatter plot graph
plt.scatter(x,y, label='Scatter Plot', color='red', marker='o')

#adding labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter plot Example')

#display the legend
plt.legend()

#display the plot 
plt.show()
```

#### 4.Histograms

```python
import numpy as np

#generate random data for histogram
data=np.random.randn(1000)

#create a histogram
plt.hist(data,bins=30,edgecolor='black',alpha=0.8,color='orange',label='Histogram')

#adding labels and titles
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('HIstogram Example')

#display the legend
plt.legend()

#display the plot 
plt.show()
```

#### 5.pie chart / plot

### 1.Colors,Markers and Linestyles:

```python
#sample data
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[1,3,5,7,9]

#customizing a line plot
plt.plot(x,y1, label='Line1', color='blue',linestyle='--',marker='o',markersize='6')
plt.plot(x,y2, label='Line2', color='yellow',linestyle='-',marker='s',markersize='7')
#adding labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Customized Multiple Line Plot ')

#display the legend
plt.legend()

#display the plot 
plt.show()
```

### 2.Annotation and Text

```python
#sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

#adding annotations and text
plt.plot(x,y, label='Line Plot',color='green',marker='o')
plt.annotate('max value',xy=(4, 8),xytext=(3, 9),
            arrowprops=dict(facecolor='black',shrink=0.06))

#adding labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('PLot with Annotation')

#display the legend
plt.legend()

#display the plot 
plt.show()
```

### Object Oriented Interface

#### Figure class

```python
#python program to show pyplot module
import matplotlib.pyplot as plt

#creating a new figure
fig = plt.figure(figsize=(4,2))

#creating new axis for the figure
ax = fig.add_axes([1,1,1,1])
x=[2,3,4,5,5,6,6]
y = [5,7,1,3,4,6,8]

ax.plot(x,y)
plt.show()
```

### adding grid lines

```python
plt.plot([1,2,3])
plt.grid(color='gray',linestyle='--',linewidth=0.7)
plt.show()
```

### multiplots

#### method 1: using add_axes()

```python
fig =plt.figure(figsize=(5,4))

ax1=fig.add_axes([0.1,0.1,0.8,0.8])

ax2=fig.add_axes([0.5,0.5,0.3,0.3])

ax1.plot([5,4,3,2,1],[2,3,4,5,6])
ax2.plot([1,2,3,4,5],[2,3,4,5,6])
plt.show()
```

#### method2:using subplot()

```python
x=[3,1,3]
y=[3,2,1]
z=[1,3,1]

plt.subplot(221)
plt.plot(x,y)

plt.subplot(222)
plt.plot(z,y)

plt.subplot(223)
plt.plot(z,y)
```

### method3: using subplots()

```python
fig,axes=plt.subplots(1,2)

axes[0].plot([1,2,3,4],[1,2,3,4])

axes[0].plot([1,2,3,4],[4,3,2,1])

axes[1].plot([1,2,3,4],[1,1,1,1])
```

### Setting Limits and Tick labels

```python
x=[3,1,3]
y=[3,2,1]

fig = plt.figure(figsize=(5,4))

ax=fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,y)
ax.set_xlim(1,2)
ax.set_xticklabels(('one','two','three','four','five','six'))
plt.show()
```

### 3D plot

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]

fig=plt.figure()

ax=plt.axes(projection='3d')
ax.plot3D(z,y,x)
```

### Area plot/stack plot

```python
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
 
plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
```
