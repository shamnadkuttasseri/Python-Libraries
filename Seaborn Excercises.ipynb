```python
1.What is Seaborn?
```

Seaborn is a Python data visualization library built on top of matplotlib. It provides a high-level interface for creating 
informative and visually appealing statistical graphics. Seaborn is designed to work seamlessly with pandas data structures
and complements the functionality of matplotlib by providing easier syntax and more advanced statistical plotting capabilities.

```python
3.What is the Seaborn function for colouring plots?
```

1.Color palettes: Seaborn offers a variety of built-in color palettes that you can use to customize the colors of your plots.
You can specify a palette using the "palette" parameter in most plotting functions.

2.Color maps: For plots that involve continuous data, such as heatmaps or contour plots, you can use color maps (colormaps) to map numeric values to colors. Seaborn provides various colormaps that you can specify using the "cmap" parameter.

3.Custom colors: You can also specify custom colors using hexadecimal color codes or named colors in functions that accept a "color" parameter.

```python
import matplotlib.pyplot as plt 

taxis=sns.load_dataset("taxis")
taxis.head()
```

```python
sns.scatterplot(x="fare", y="tip", hue="color", palette="Set1", data=taxis)
plt.show()
```

```python
sns.scatterplot(x="fare", y="tip", color="orange", data=taxis)
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights_pivoted=flights.pivot(index="month",columns="year",values="passengers")

ax=sns.heatmap(flights_pivoted,cmap="YlGnBu")
```

```python
4.How do you plot a histogram in Seaborn?
```

```python
sns.histplot(data=taxis,x="fare",bins=40)
plt.show()
```

```python
sns.histplot(data=taxis,x="fare",kde=True)
plt.show()
```

```python
5.How to change the minimum number of ticks of MaxNLocator in Seaborn?
```

```python

sns.scatterplot(x='fare', y='tip', data=taxis)

ax = plt.gca()

ax.xaxis.get_major_locator().set_params(nbins=8) 
plt.show()
```

```python
6.How to color the data points by a category like to assign colors to the 'regional indicators'?
```

```python
sns.scatterplot(x='fare', y='tip', data=taxis,hue="payment")
plt.show()
#to change the color
sns.scatterplot(x='fare', y='tip', data=taxis,hue="payment",palette="Set1")
plt.show()
```

```python
7.How do you comparev all the continuous features in a dataset using seaborn?
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

taxis=sns.load_dataset("taxis")
sns.pairplot(taxis,hue="total")

plt.show()
```

```python
8.How will you plot Categorical data?
```

```python
sns.catplot(data=taxis, kind="strip", x="fare", y="tip", hue="color")

plt.show()
```

```python
10.How can you can draw relationships between two continuous variables?
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='fare', y='tip', data=taxis, hue="payment", label="scatter plot")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Relationship between X and Y')

plt.legend()

plt.show()
```
