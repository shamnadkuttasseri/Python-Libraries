```python
6.Create a stacked bar chart that shows the total number of cars sold by month, broken down by brand. Use the data provided 
below:
Month	Toyota	Honda	Ford
Jan	120	90	80
Feb	100	110	90
Mar	140	80	100
Apr	90	120	110
May	110	100	120
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
toyota = [120, 100, 140, 90, 110]
honda = [90, 110, 80, 120, 100]
ford = [80, 90, 100, 110, 120]

fig, ax = plt.subplots()
ax.bar(months, toyota, label='Toyota')
ax.bar(months, honda, bottom=toyota, label='Honda')
ax.bar(months, ford, bottom=[toyota[i] + honda[i] for i in range(len(months))], label='Ford')

ax.set_ylabel('Number of Cars Sold')
ax.set_xlabel('Month')
ax.set_title('Total Number of Cars Sold by Month and Brand')
ax.legend()

plt.show()
```

```python
7.Create a scatter plot of the iris dataset, with the x-axis representing the sepal length and the y-axis representing the
petal length. Color the points according to the species of iris.
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

sepal_length = X[:, 0]
petal_length = X[:, 2]

plt.figure(figsize=(8, 6))
plt.scatter(sepal_length, petal_length, c=y, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot of Iris Dataset (Sepal Length vs. Petal Length)')
plt.colorbar(label='Species')
plt.show()
```

```python
8.Create a line chart that shows the number of daily visitors to a website over a period of one month. Use the random 
module to generate some fake data for the number of visitors.
```

```python
import random

num_days = 30
daily_visitors = [random.randint(100, 500) for _ in range(num_days)]

days_of_month = range(1, num_days + 1)

plt.figure(figsize=(10, 6))
plt.plot(days_of_month, daily_visitors, marker='o', linestyle='-')
plt.title('Number of Daily Visitors to Website')
plt.xlabel('Day of the Month')
plt.ylabel('Number of Visitors')
plt.grid(True)
plt.xticks(range(1, num_days + 1))  # Show all days on x-axis
plt.tight_layout()

plt.show()
```

```python
9.. Create a stacked area chart that shows the percentage of time a person spends on different activities during the day. Use the data provided below:
Time of Day	Work	Exercise	Leisure
Morning	20%	5%	75%
Afternoon	60%	10%	30%
Evening	30%	5%	65%
```

```python
time_of_day = ['Morning', 'Afternoon', 'Evening']
work = [20, 60, 30]
exercise = [5, 10, 5]
leisure = [75, 30, 65]

plt.figure(figsize=(10, 6))
plt.stackplot(time_of_day, work, exercise, leisure, labels=['Work', 'Exercise', 'Leisure'], alpha=0.8)

plt.xlabel('Time of Day')
plt.ylabel('Percentage of Time')
plt.title('Percentage of Time Spent on Different Activities During the Day')
plt.legend(loc='upper left')

plt.show()
```

```python
10.Create a 3D scatter plot of random points, where the x, y, and z values are all generated randomly using NumPy's random 
module. Add labels to the x, y, and z axes.
```

```python
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='b', marker='o')

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

ax.set_title('3D Scatter Plot of Random Points')

plt.show()
```

```python
11.Create a stacked bar chart that shows the total number of hours spent by employees on different projects during a week. 
Use the data provided below:
Project	Employee A	Employee B	Employee C
Project 1	10	5	15
Project 2	5	10	20
Project 3	15	5	10
```

```python
projects = ['Project 1', 'Project 2', 'Project 3']
employees = ['Employee A', 'Employee B', 'Employee C']
hours_spent = {
    'Employee A': [10, 5, 15],
    'Employee B': [5, 10, 5],
    'Employee C': [15, 20, 10]
}

fig, ax = plt.subplots(figsize=(10, 6))

for i, employee in enumerate(employees):
    if i == 0:
        ax.bar(projects, hours_spent[employee], label=employee)
    else:
        bottom = [sum(hours_spent[employees[j]][:k+1]) for k in range(len(projects))]
        ax.bar(projects, hours_spent[employee], bottom=bottom, label=employee)

ax.set_xlabel('Projects')
ax.set_ylabel('Hours Spent')
ax.set_title('Total Hours Spent by Employees on Different Projects')
ax.legend()

plt.show()
```
