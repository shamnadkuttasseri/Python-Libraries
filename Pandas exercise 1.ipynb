```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head(5)
```

```python
#2.How many passengers were in first class (Pclass=1) in the titanic dataset?
```

```python
df.columns
```

```python
df1=df.pclass.value_counts()

counts_firstclass_passengers=df1.loc[1]
print("number of first class passengers:")
print(counts_firstclass_passengers)
```

```python
# 3.Create a new column in the titanic dataset called "is_alone" which should be True if a passenger has no siblings/spouses
# or parents/children on board and False otherwise.
```

```python
df['is_alone']=pd.Series(df['sibsp']+df['parch']==0)

df.head(5)
```

```python
#4.Group the titanic dataset by "Pclass" and calculate the mean "Age" for each class.
```

```python
group_by_class=df.groupby('pclass')

for class_name, group_data in group_by_class:
    print("Class:", class_name)
    print(group_data.head())
```

```python
mean_age_class=group_by_class['age'].mean()

print("mean age for each class:")
print(mean_age_class)
```

```python
#5.Sort the titanic dataset by "Fare" in descending order and display the first 10 rows.
```

```python
sorting_by_fare=df.sort_values(by='fare', ascending=False)
sorting_by_fare.head(10)
```

```python
# 6.Create a new column in the titanic dataset called "fare_category" which should have the value "cheap" if the fare is less 
# than 10, "moderate" if the fare is between 10 and 50, and "expensive" if the fare is greater than 50.
```

```python
import numpy as np

conditions = [
    df['fare'] < 10,
    (df['fare'] >= 10) & (df['fare'] <= 50),
    df['fare'] > 50
]

categories = ['cheap', 'moderate', 'expensive']

df['fare_category'] = np.select(conditions, categories, default='Unknown')

df.head()
```

```python
#7.Filter the titanic dataset to only include passengers who are male and survived.
```

```python
filtered_df=df[(df['sex']=='male')&(df['survived']==1)]

filtered_df.head()
```
